```python
from roboflow import Roboflow
rf = Roboflow(api_key="your api key")
project = rf.workspace("aamir-naseer-widhl").project("fire-and-smoke-5")
version = project.version(1)
dataset = version.download("yolov8")          
```

```python
import matplotlib
matplotlib.use('Agg')  # Use non-GUI backend suitable for scripts
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data=f'{dataset.location}/data.yaml',  # Path to your dataset config
    epochs=1,
    imgsz=416,
    batch = 16,
    conf=0.25,
    iou=0.5,
    device='mps'
)
```

```python
model.save("./yolov8n-fire-init.pt")
```

```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

images_dir = os.path.join(dataset.location, "valid", "images")
image_files = [f for f in os.listdir(images_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

random_image_file = random.choice(image_files)
random_image_path = os.path.join(images_dir, random_image_file)

image = Image.open(random_image_path)

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image

results = model.predict(source=random_image_path, device='mps', save=False, show=False, stream=True, conf=0.25, iou=0.5)

result = next(results)

img_with_boxes = result.plot()

img_rgb = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
